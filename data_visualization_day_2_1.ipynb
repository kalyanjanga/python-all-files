##### import the titanic dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"titanic.csv")
data.head()
```

##### 1. Write a Python program to display a bar chart of the fare of class.

```python
sns.barplot(x="class",y="fare",data=data,color="b")
```

###### 2. Write a Python program to display a horizontal bar chart of the fare of class.

```python
sns.barplot(y="class",x="fare",data=data,color="b")
```

###### 3. Write a Python program to display a bar chart of the fare of class. Use different color for each bar.

```python
sns.barplot(x="class",y="fare",data=data)
```

###### 4. Write a Python program to create a pie chart of the fare of class.

```python
FC = pd.pivot_table(data=data,index="class",values="fare")
plt.pie(FC.fare,labels=["First","Second","Third"],autopct="%0.1f%%")
plt.title("Fare of class")
plt.show()
```

##### 5. Write a Python program to display a count plot of all categorical columns.

```python
# Count plot for all categorical columns
plt.figure(figsize=(10,5))
plt.subplot(3,3,1)
sns.countplot(x="sex",data=data)
plt.subplot(3,3,2)
sns.countplot(x="embarked",data=data)
plt.subplot(3,3,3)
sns.countplot(x="class",data=data)
plt.subplot(3,3,4)
sns.countplot(x="who",data=data)
plt.subplot(3,3,5)
sns.countplot(x="adult_male",data=data)
plt.subplot(3,3,6)
sns.countplot(x="deck",data=data)
plt.subplot(3,3,7)
sns.countplot(x="embark_town",data=data)
plt.subplot(3,3,8)
sns.countplot(x="alive",data=data)
plt.subplot(3,3,9)
sns.countplot(x="alone",data=data)
plt.tight_layout()
```
